```python
def build_poly_regressors(x, poly_order=1):
    
    poly_regressors = np.hstack((np.ones((x.shape[0], 1)), x))
    
    if poly_order > 1:
        for i in range(2,poly_order+1):        
            poly_regressors = np.hstack((poly_regressors, x**i))
            
    return poly_regressors
```
